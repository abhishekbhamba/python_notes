```python
s = "Python Programming"
```

```python
s[::2]         # step size is 2.
               # it go in positive direction.
```

```python
s[::-2]       # step size is -2.
              # it will start with negative direction.
```

```python
s[2:3]
```

```python
s[2:4]      # basically it will not count the last element in python.           
```

```python
s[2:3:-1]     # here 2:3 is going in positive direction.
              # but -1 is showing negative direction.
```

```python
s[-15:-16]        # it is giving me empty set 
                   # because from -15 to -16 is going in negative direction
```

```python
s[-15:-16:-1]       # so, the last value could not count.
```

```python
s[-6:7:-2]          # so, it will start with -6 and go until the 7 index no.
                    # the i want step size -2.
```

```python
s[4:-8]       # it is posssible from -4 to 8
```

```python
s[-4:8]          # but from -4 to 8 it is not possible.
```

```python
s[9:-10]        # it is giving empty list 
```

```python
s[-9:9]
```

```python
s[9:-10:-1]
```

```python
s[22:-6]         # it will give empty set.
                # because 22 is extreme no. or it exceed the llimit. 
```

```python
s[22::-6]
```

```python
s[2:-22:-1]
```

```python
s[15::-2]         # -2 shows negative relation.
```

```python
s[-15::-2]
```
